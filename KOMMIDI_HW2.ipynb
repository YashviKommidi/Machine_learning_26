#### Question 1

For each of four scenarios listed below, indicate whether we would generally expect the performance of a flexible statistical learning method to be better or worse than an inflexible method. Justify your answer.

- The sample size (number of observations; n) is extremely large, and the number of predictors (features; p) is small.

> **_Answer:_**  when we have a lot of data as the sample size, but only few features, the flexible method will perform better. Large sample size generally means that we have enough data to accurately estimate a complex model without worrying much about overfitting. Whereas with few  features, we are not dealing with the curse of dimensionality, so with the flexible method we can learn the true patterns in the data. There is enough data to support any complexity that comes with the flexibility.

- The number of predictors (p) is extremely large, and the number of observations (n) is small.

> **_Answer:_**  When the feature count is extremely large with a small count of observations, flexible statistical method is worse as those models can overfit the training data. As the dataset is small, models can find patterns that are just noise as it can be difficult to distinguish the real patterns from the data. An inflexible method could be a better option here like linear regression because it has fewer parameters to estimate, so it requires less data and is less prone to overfitting. In this scenario, simpler models generalize better to the new data.

- The relationship between the predictors and response is highly non-linear.

> **_Answer:_** When the relationship between the features and the response is highly non-linear, flexible statistical learning methods perform better. As inflexible methods like linear regression assume a straight-line relationship, they will have high bias when the actual relationship is a non-linear or curved one . In this case, flexible methods can adjust to the data, that is they can match to the actual shape of the data. As the relationship is non-linear, flexible methods will have a much lower bias and will help to capture the real patterns which in general a linear method would have missed to identify.

- The variance of the error terms (i.e. $σ^2 = Var(ϵ)$) is extremely high.

> **_Answer:_**  When the variance of the error terms is extremely high, a flexible method will perform worse than an inflexible method. Generally high error variance means there is a lot of irreducible noise in the data that no model can usually predict. It is known that flexible methods generally have high variance and it will try to fit every fluctuation in the data including the noise. This leads to overfitting. Whereas an inflexible method is more rigid and usually won't chase the random noise as much as flexible method does. So it will have higher bias but lower variance, which is better suitable when the data is highly noisy. Since we cannot reduce the irreducible error anyway, using a simpler model helps to prevent us from making predictions worse by fitting the noise.

- Briefly explain the “curse of dimensionality”, provide a hypothetical example illustrating the concept, and list at least one potential way that is generally used to handle it when using machine learning models.

> **_Answer:_**  The curse of dimensionality is a kind of problem that is seen generally when we are working with higher-dimensional data. Usually, the column of the feature space grows as the number of features increases, which makes it hard to find data, as it is scattered. Because of this, we can know that the data points become farther apart from each other, and it makes it hard for the models that rely on measuring distances or finding the nearest data points.

> for example, when we are trying to apply for an internship, we will begin with internships that are active and in tucson, here it is just 2 features, location and status, then later we go on increase features like skills, technologies, experience required, domain specific openings, time when the job is opened, company it belongs to, what is the visa status that is needed to be eligible, and many more features. As we keep on increasing the features, at some point, there are almost no job matches to the requirement on all the dimensions. The data becomes too spread out, and finding a truly similar job that we are searching for becomes nearly impossible, even with jobs across the country.

> one of the common methods used to handle the curse of dimensionality is dimensionality reduction or feature selection. With the name itself, we can understand that, in this method, we will be reducing the number of dimensions, that is, features, by keeping only the most important or just combining multiple related features. For example, Principal component analysis is one of the techniques that can transform many features into fewer new features that capture most of the important information. This makes the data less sparse and helps the models work better.

- Explain how dimensionality increases in each of the following cases and describe the consequences for learning:
    - Adding Features: Increasing the number of input features while keeping the model linear.
    - Polynomial Expansion: Increasing the polynomial degree of a model while keeping the original input features fixed.

> **_Answer:_**  When we add more input features to a linear model, the dimensionality increases directly, as each new feature adds one more dimension to the feature space. As the dimension count increases, the data becomes sparser as it spreads out in higher dimensions. This situation can cause overfitting, especially if the number of features is close to the number of observations. Here, the model will have a higher number of parameters to estimate, so it needs more data to learn the patterns. With the limited data, the model might fit the noise hidden in the data instead of the true patterns.
When we increase the polynomial degree, we are creating new features from the existing features. For example, with features X1 and X2, degree 2 creates X12, X22, and X1X2. The number of features grows rapidly with the polynomial degree. So with higher polynomial degrees, dimensionality dramatically increases. This leads to overfitting because the model becomes very flexible and can fit complex patterns in the training data, including noise. The model will also have high variance and poor generalization to test data. Along with that, it also increases the computational costs and makes the model harder to interpret. Without enough training data, high-degree polynomial models will memorize the training data rather than learn meaningful patterns.

So training error always goes down as the model complexity is increased. When it comes to test error, this behaves completely different. As at first, the test error decreases as we add complexity, because as the model is getting better, test error decreases. But at certain point, the model becomes too complex and it starts fitting the noise in the training data instead of the true patterns. This is when the test error starts increasing. Because when we test this model on the new data, those noise patterns doesn't exist, which leads to poor performance. 
These differences are seen because the training data is what the model actually sees and learns from, so it always fits better as the complexity increases , which reduces the training error. The right amount of the complexity is when the test error is the lowest and before it starts increasing.

- To the best of your knowledge, distinguish between training, test, and validation datasets. Briefly describe the importance of each in the context of Machine Learning.

> **_Answer:_**  To deal with a machine learning model, the whole dataset is generally divided into 3 sets, which is training, test and validation set. Training set is the set of data that we use to train our model and help the model to learn about the parameters. This data is exposed to the model, that is whole training data is in the memory of the model. When it comes to validation set, it is used during the model development to evaluate and compare different models or hyperparameter settings. This data is generally used for tuning the hyperparameters, helps to select the best model and its suitable parameters. This will also help the models to avoid overfitting
And the test data is the data which we use to test the model, this data is completely new to the model. We use this data after training the model, to evaluate the performance of the model. This data helps to detect if the model has learned patterns specific to the training set or general patterns. 
Using different subsets of the data helps us to build models that generalize well to new data rather than just memorizing the training data and performing accordingly.
