Suppose that a curve $\hat{g}$ is computed to smoothly fit a set of `n` points using the following formula:

  $$
  \hat{g} = \arg\min_{g}(\sum_{i=1}^n (y_i - g(x_i))^2 +  \lambda \int [g^{(m)} (x)]^2 dx )

This question relates to the `College` data set.

a) Split the data into a training set and a test set (70:30). Using out-of-state tuition as the response and the other variables as the predictors, perform *forward* stepwise selection on the training set in order to identify a satisfactory model that uses just a subset of the predictors.

```python
#importing libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

#loading the data and preparing the data
df = pd.read_csv('College.csv', index_col=0)
df['Private'] = (df['Private'] == 'Yes').astype(int)

X = df.drop('Outstate', axis=1)
y = df['Outstate']

#performing the 70:30 split for the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# doing the forward stepwise selection
remaining, selected, results = list(X.columns), [], []

while remaining:
    best_rss, best_feat = np.inf, None
    for feat in remaining:
        model = LinearRegression().fit(X_train[selected + [feat]], y_train)
        rss = mean_squared_error(y_train, model.predict(X_train[selected + [feat]])) * len(y_train)
        if rss < best_rss:
            best_rss, best_feat = rss, feat
#adding the best predictor to selected and remove from remaining
    selected.append(best_feat)
    remaining.remove(best_feat)
    results.append((len(selected), selected.copy(), best_rss))

#picking the best model using 5-fold cross validation
best_cv, best_feats = np.inf, None
for _, feats, _ in results:
    cv = -cross_val_score(LinearRegression(), X_train[feats], y_train,
                          cv=5, scoring='neg_mean_squared_error').mean()
    if cv < best_cv:
        best_cv, best_feats = cv, feats

#printing the best features and the CV MSE
print(f"Best features are :({len(best_feats)}): {best_feats}")
print(f"CV MSE: {best_cv:.2f}")
```

b) Fit a GAM on the training data, using out-of-state tuition as the response and the features selected in the previous step as the predictors. Plot the results, and briefly explain your findings.

```python

from pygam import LinearGAM, s
import matplotlib.pyplot as plt

#fitting the GAM model using the selected features
X_train_gam = X_train[best_feats].values

# building the GAM with a spline for each predictor
gam = LinearGAM(s(0) + s(1) + s(2) + s(3) + s(4) + s(5) + s(6) +
                s(7) + s(8) + s(9) + s(10) + s(11) + s(12))

#using the gridsearch to automatically find the best smoothing parameter lambda
gam.gridsearch(X_train_gam, y_train)

#plotting the results we get
fig, axes = plt.subplots(4, 4, figsize=(16, 12))
axes = axes.flatten()

#looping through each of the features and also plotting its partial dependence
for i, feat in enumerate(best_feats):
    XX = gam.generate_X_grid(term=i)
    axes[i].plot(XX[:, i], gam.partial_dependence(term=i, X=XX))
    axes[i].plot(XX[:, i], gam.partial_dependence(term=i, X=XX, width=0.95)[1],
                 c='r', ls='--', alpha=0.5)
    axes[i].set_title(feat)
    axes[i].set_xlabel(feat)
    axes[i].set_ylabel('Effect on Outstate')

plt.tight_layout()
plt.show()
```

  b) Now, fit a Poisson GAM that predicts richness using year and location. Note that location is actually two features: `Longitude` and `Latitude`. Name this model `M1`. Plot the model, get the `summary` of `M1`, and answer: Is there evidence for effects of location and year?

```python
from pygam import PoissonGAM, s, te
import matplotlib.pyplot as plt

#loading the data
df = pd.read_csv('HW6_data', sep='\t', index_col=0)
X = df[['Year', 'Latitude', 'Longitude']].values
y = df['richness'].values

#then fitting the Poisson GAM M1 that is s() for year, te() for location jointly
M1 = PoissonGAM(s(0) + te(1, 2))
M1.gridsearch(X, y)

#printing the summary of M1
print(M1.summary())

#plotting the results of M1
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

#generating the plots
XX = M1.generate_X_grid(term=0)
axes[0].plot(XX[:, 0], M1.partial_dependence(term=0, X=XX))
axes[0].plot(XX[:, 0], M1.partial_dependence(term=0, X=XX, width=0.95)[1], c='r', ls='--', alpha=0.5)
axes[0].set_title('effect of year')
axes[0].set_xlabel('year')

XX = M1.generate_X_grid(term=1)
axes[1].plot(XX[:, 1], M1.partial_dependence(term=1, X=XX))
axes[1].plot(XX[:, 1], M1.partial_dependence(term=1, X=XX, width=0.95)[1], c='r', ls='--', alpha=0.5)
axes[1].set_title('effect of location')
axes[1].set_xlabel('latitude/longitude')

plt.tight_layout()
plt.show()
```

from the summary we can say that there is strong evidence for both location and the year. for the year s(0) has a p value of 0.00 (***) with EoDF of 20 which confirms us a strong and highly non linear temporal effect on richness. When it comes to Location, te(1,2) has a p value of 0.00(***) with EoDF 54.2 which confirms us a strong spatial effect. The Pseudo value R² = 0.53 which means that the model explains about 53% of the variance in richness using just year and location. we have used te() for location because latitude and longitude interact jointly to determine location and cannot be treated independently as separate predictors.
