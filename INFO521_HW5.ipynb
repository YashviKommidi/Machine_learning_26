#### Question 3

a) What is the following code supposed to do?

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.preprocessing import PolynomialFeatures
import statsmodels.api as sm

# Simulate some data
np.random.seed(0)
n = 100
order = 2
x = np.random.uniform(-4, 4, n)
y = np.random.normal(loc=x**order, scale=0.25, size=n)
data = pd.DataFrame({'x': x, 'y': y})

# Number of folds
n_folds = 20

# Polynomial degrees
poly_degrees = range(1, 11)

# KFold cross-validation
kf = KFold(n_splits=n_folds, shuffle=True, random_state=1)
mse_estimates = []

# Loop over polynomial degrees
for q in poly_degrees:
    y_hat = np.zeros(n)
    
    for train_idx, test_idx in kf.split(data):
        # Fit on K-1 folds
        poly = PolynomialFeatures(degree=q)
        X_train = poly.fit_transform(data.iloc[train_idx][['x']])
        X_test = poly.transform(data.iloc[test_idx][['x']])
        
        ols = sm.OLS(data.iloc[train_idx]['y'], X_train).fit()
        
        # Predict on the Kth fold
        y_hat[test_idx] = ols.predict(X_test)
    
    # Get the MSE estimate
    mse = np.mean((data['y'] - y_hat) ** 2)
    mse_estimates.append(mse)
```

What is the best polynomial order (1–20) modeling the structure of the following dataset (`x` vs `y`)? Your answer should include a numerical approach (e.g. MSE estimates) along with a convincing set of visuals supporting your answer.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(4)
n = 1000
x = np.random.uniform(-4, 4, n)
mu = np.where(x > 0, np.cos(4 * x), 1 - np.sin(x))
y = mu + np.random.normal(size=n)
data = pd.DataFrame({'x': x, 'y': y})

plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='black', edgecolor='steelblue', s=50, label='Data', alpha=0.7)
x_ordered = np.sort(x)
mu_ordered = mu[np.argsort(x)]

plt.plot(x_ordered, mu_ordered, color='red', linewidth=3, label="E[Y|X=x], truth")

plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc='lower left')
plt.show()
```
