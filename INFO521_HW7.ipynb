```python
import pandas as pd
import statsmodels.api as sm
from sklearn.metrics import confusion_matrix

Data1Raw = pd.read_csv('sick.data', sep = '\t')
Data1 = Data1Raw.drop(columns=['TBG', 'TBG_measured', 'hypopituitary', 'TSH', 'T3', 'TT4', 'T4U', 'FTI']).dropna()
Data1['Class'] = Data1['Class'].map({'sick': 1, 'negative': 0})

binary_cols = Data1.columns[Data1.isin(['f', 't']).all()]
Data1[binary_cols] = Data1[binary_cols].map(lambda x: 1 if x == 't' else 0)
Data1['sex'] = Data1['sex'].map({'F': 0, 'M': 1})
Data1 = pd.get_dummies(Data1, columns=['referral_source'], dtype='int')

X = Data1.drop(columns=['Class'])
y = Data1['Class']
X = sm.add_constant(X)
glm_model = sm.GLM(y, X, family=sm.families.Binomial())
glm_results = glm_model.fit()

y_pred = glm_results.predict(X)
y_pred_class = [1 if p > 0.5 else 0 for p in y_pred]
conf_matrix = confusion_matrix(y, y_pred_class)
print(conf_matrix)
```

##### Data pre-processing

```python
#importing libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

#loading the sick data set 
sick_rawdata = pd.read_csv('sick.data', sep='\t')
print(sick_rawdata.shape)
print(sick_rawdata['Class'].value_counts())

# dropping the columns TBG, TBG_measured, and hypopituitary
# dropping TBG becuase the values are 100% missing 
# dropping TBG_measured because all the values are always f
# dropping hypopituitary because the values has 3771 f and 1 t
sick_data = sick_rawdata.drop(columns=['TBG', 'TBG_measured', 'hypopituitary'])

#adding the filter for age as more than 120 would be practically impossible
sick_data = sick_data[sick_data['age'] <= 120]

# encoding the target 
sick_data['Class'] = sick_data['Class'].map({'sick': 1, 'negative': 0})

#encoding binary t/f columns to 1/0
binary_cols = [col for col in sick_data.columns if set(sick_data[col].dropna().unique()).issubset({'t', 'f'})]
for col in binary_cols:
    sick_data[col] = sick_data[col].map({'t': 1, 'f': 0})

# encoding the sex column
#imputing the 150 missing sex values with mode which is the most frequent = F before encoding
sick_data['sex'] = sick_data['sex'].fillna('F')
sick_data['sex'] = sick_data['sex'].map({'F': 0, 'M': 1})

# converting the categories into numbers so the model can understand them
sick_data = pd.get_dummies(sick_data, columns=['referral_source'], drop_first=True, dtype=int)

#separating the features and target
X = sick_data.drop(columns=['Class'])
y = sick_data['Class']

# imputing the missing values with median
imputer = SimpleImputer(strategy='median')
X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

# splitting the data into Train and test
X_train, X_test, y_train, y_test = train_test_split( X_imputed, y, test_size=0.2, random_state=42, stratify=y )

#scaling the features

#  fixing  the overflow warning seen in the collaborator's code
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)

#printing results
print(f"Train: {X_train.shape[0]} rows | Test: {X_test.shape[0]} rows")
```

##### Model development

```python
#collaborator's code
import pandas as pd
import statsmodels.api as sm
from sklearn.metrics import confusion_matrix, recall_score

Data1Raw = pd.read_csv('sick.data', sep = '\t')
Data1 = Data1Raw.drop(columns=['TBG', 'TBG_measured', 'hypopituitary', 'TSH', 'T3', 'TT4', 'T4U', 'FTI']).dropna()
Data1['Class'] = Data1['Class'].map({'sick': 1, 'negative': 0})

binary_cols = Data1.columns[Data1.isin(['f', 't']).all()]
Data1[binary_cols] = Data1[binary_cols].map(lambda x: 1 if x == 't' else 0)
Data1['sex'] = Data1['sex'].map({'F': 0, 'M': 1})
Data1 = pd.get_dummies(Data1, columns=['referral_source'], dtype='int')

X = Data1.drop(columns=['Class'])
y = Data1['Class']
X = sm.add_constant(X)
glm_model = sm.GLM(y, X, family=sm.families.Binomial())
glm_results = glm_model.fit()

y_pred = glm_results.predict(X)
y_pred_class = [1 if p > 0.5 else 0 for p in y_pred]
conf_matrix = confusion_matrix(y, y_pred_class)
print("collaborator's original results")
print(conf_matrix)
print(f"Recall: {recall_score(y, y_pred_class):.3f}")
```

```python
#importing libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, recall_score, classification_report

# the model i am using is Random Forest
# using random forest because it handles complex relationships between hormone features better than logistic regression
rf_model = RandomForestClassifier(class_weight='balanced', random_state=42)
rf_model.fit(X_train_scaled, y_train)

# using threshold of 0.3 to catch more sick patients
rf_probs = rf_model.predict_proba(X_test_scaled)[:, 1]
rf_preds = (rf_probs >= 0.3).astype(int)

print("random forest results are")
print(confusion_matrix(y_test, rf_preds))
print(classification_report(y_test, rf_preds))
print(f"Recall: {recall_score(y_test, rf_preds):.3f}")

# comparing collaborator's model vs our random forest
models = ['collaborator', 'random forest']
recalls = [0.813, recall_score(y_test, rf_preds)]

plt.bar(models, recalls, color=['tomato', 'steelblue'])
plt.title('recall comparison')
plt.ylabel('recall score')
plt.ylim(0.7, 1.0)
plt.show()
```
