#### Question 4
Randomly sample 20 elements (with replacement) 100 times from *Q3*. 
Save the results in a new variable named *Q4* as a list.

```python
# here we are sampling 20 elements from q3 with replacement and storing them in Q4
Q4 = [np.random.choice(Q3, size=20, replace=True).tolist() for _ in range(100)]
print(Q4)
```

```python
#another way to solve the above question 
 #Initializing the empty list
Q4 = [] 

#using for loop, getting 
for i in range(100):
    # Sampling 20 elements from Q3 with replacement and converting into list and storing in single_sample 
    single_sample = np.random.choice(Q3, size=20, replace=True).tolist()
    # appending each sample to the Q4
    Q4.append(single_sample)

print(Q4)
```
