```python
# importing the libraries
import matplotlib.pyplot as plt
import seaborn as sns

#selecting only the numeric columns for correlation
numeric_data = weekly_data.select_dtypes(include=[np.number])

# printing the correlation between the predictors
print(numeric_data.corr())

# printing the graphical summary that is orrelation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('correlation heatmap of weekly dataset')
plt.show()
```

  b) Use the full data set to perform a logistic regression with `Direction` as the response and the five `lag` variables plus `Volume` as predictors. Print the regression results. Do any of the predictors appear to be statistically significant? If so, which ones? 

```python

import statsmodels.api as sm

# defining the predictors and the response variable
X = weekly_data[['Lag1', 'Lag2', 'Lag3', 'Lag4', 'Lag5', 'Volume']]
y = (weekly_data['Direction'] == 'Up').astype(int)

#adding the constant to the predictors
X_sm = sm.add_constant(X)

#fitting the logistic regression model
log_reg = sm.Logit(y, X_sm)
result = log_reg.fit()

#printing the regression results
print(result.summary())
```

from the results we got from the above that is the regression results, we can see that the only Lag2 appears to be statistically significant because it has a p-value of 0.030 which is less than 0.05. This generally means that Lag2 has a positive relationship with the direction of. the market, which means that if the market went up two weeks ago, then it is more likely to go up this week as well. When it comes to all the other predictors such as Lag1, Lag2, Lag3, Lag4, Lag5 and volume do not appear to be statistically significant as their p values are greater than 0.05. so we can say that the previous week returns except for the lag2 do not have a strong effect on the current week;

c) Compute the __confusion matrix__ and overall fraction of correct predictions. Explain what the confusion matrix is telling you about the types of mistakes made by logistic regression.

```python
#importing libraries
from sklearn.metrics import confusion_matrix, accuracy_score

# getting the predicted values from the model we fitted in before question
y_pred = result.predict(X_sm)

#converting the predicted probabilities to 0 and 1
y_pred_class = (y_pred >= 0.5).astype(int)

# computing the confusion matrix
cm = confusion_matrix(y, y_pred_class)
print("confusion matrix:")
print(cm)

# computing the overall fraction of correct predictions
accuracy = accuracy_score(y, y_pred_class)
print(f"overall fraction of correct predictions: {accuracy:.4f}")
print(f"value in percentage: {accuracy*100:.2f}%")
```

Develop a model to predict whether a given car gets high or low gas mileage based on the `Auto` data set. Load the dataset:

```python
import pandas as pd
import numpy as np

# loading the auto dataset
auto_data = pd.read_csv('Auto.data', sep='\s+', na_values='?')

# dropping if there are any missing values
auto_data = auto_data.dropna()

#printing few rows
print(auto_data.head())
```

a) Create a binary variable, `mpg01`, that contains a `1` if `mpg` contains a value above its median, and a `0` if `mpg` contains a value below its median. You can compute the median using the `median()` function.

```python
# computing the median of mpg to use for binary variable
mpg_median = auto_data['mpg'].median()
print(f"median of mpg is {mpg_median}")

# creating the binary variable mpg01
# condition which is given in question is to have 1 if mpg is above the median, 0 if mpg is below the median
auto_data['mpg01'] = (auto_data['mpg'] > mpg_median).astype(int)

#printing the mpg01 value count
print(auto_data['mpg01'].value_counts())
print(auto_data.head())
```

b) __Explore__ the data graphically in order to investigate the association between `mpg01` and the other features. Which of the other features seem most likely to be useful in predicting `mpg01`? _Scatterplots and boxplots may be useful tools to answer this question_. __Describe your findings__.

```python
#importing libraries
import matplotlib.pyplot as plt
import seaborn as sns

# creating boxplots to look at the association between mpg01 and other features
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# printing the boxplot for cylinders
sns.boxplot(x='mpg01', y='cylinders', data=auto_data, ax=axes[0, 0])
axes[0, 0].set_title('mpg01 vs cylinders')

# printing boxplot for displacement
sns.boxplot(x='mpg01', y='displacement', data=auto_data, ax=axes[0, 1])
axes[0, 1].set_title('mpg01 vs displacement')

#printing the boxplot for horsepower
sns.boxplot(x='mpg01', y='horsepower', data=auto_data, ax=axes[0, 2])
axes[0, 2].set_title('mpg01 vs horsepower')

#printing the boxplot for weight
sns.boxplot(x='mpg01', y='weight', data=auto_data, ax=axes[1, 0])
axes[1, 0].set_title('mpg01 vs weight')

# printing boxplot for acceleration
sns.boxplot(x='mpg01', y='acceleration', data=auto_data, ax=axes[1, 1])
axes[1, 1].set_title('mpg01 vs acceleration')

# printing the boxplot for year
sns.boxplot(x='mpg01', y='year', data=auto_data, ax=axes[1, 2])
axes[1, 2].set_title('mpg01 vs year')

#showing all plots in one picture
plt.tight_layout()
plt.show()
```

c) Split the data into a training set and a test set (65%, 35%). Set the random seed to _4_ to make the training data reproducible.

```python
#importing library
from sklearn.model_selection import train_test_split


# splitting the whole data into 65% training and 35% test sets as mentioned
X = auto_data[['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'year']]
y = auto_data['mpg01']

# setting the random seed to 4 to make the training data reproducible as mentioned in question
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=4)

print(f"training set size: {X_train.shape[0]}")
print(f"testing set size: {X_test.shape[0]}")
```

Hint: Recall that the density function for a normal random variable is $f(x) = \frac{e^{-(x - \mu)^{2}/(2\sigma^{2}) }} {\sigma\sqrt{2\pi}}$. You will need to use Bayes’ theorem.

> **_Answer:_** from the question we can see that the mean for companies that issued dividend is 10 that is mu_yes, mean for the companies that did not issue dividend is 0 that is mu_no, variance is 36 which is sigma2, so the standard deviation is 6 which is sigma.

The prior probability of issuing dividend is 80% that is 0.80 which is pi_yes, the prior probability of not issuing dividend is 20% that is 0.20 which is pi_no

and atlast X=4

and from the formula $f(x) = \frac{e^{-(x - \mu)^{2}/(2\sigma^{2}) }} {\sigma\sqrt{2\pi}}$

we can calculate the density function for each class is 

f_yes = np.exp(-(X - mu_yes)**2 / (2 * sigma2)) / (sigma * np.sqrt(2 * np.pi)) 

f_no = np.exp(-(X - mu_no)**2 / (2 * sigma2)) / (sigma * np.sqrt(2 * np.pi))

then we can apply the bayes theorem that is 

P(yes|X) = (pi_yes * f_yes) / (pi_yes * f_yes + pi_no * f_no)

at last we can say that 

the probability that a company will issue a dividend this year ,given that its percentage profit was X= 4 last year is approximately 0.7519 which is 75.19%. This means that there is a  75.19% chance that the company will issue a dividend this year based on the  last year's percentage profit of X = 4.



```python
# writing in code format just in case.
import numpy as np

#values from the question
mu_yes = 10      
mu_no = 0        
sigma2 = 36      
sigma = np.sqrt(sigma2)  
pi_yes = 0.80    
pi_no = 0.20     
X = 4            

# with the help of formula we can calculate the density function for each class
f_yes = np.exp(-(X - mu_yes)**2 / (2 * sigma2)) / (sigma * np.sqrt(2 * np.pi))
f_no = np.exp(-(X - mu_no)**2 / (2 * sigma2)) / (sigma * np.sqrt(2 * np.pi))

# then we can apply the bayes theorem that is
# we know th ebayes theorem is P(yes|X) = (pi_yes * f_yes) / (pi_yes * f_yes + pi_no * f_no)
p_yes = (pi_yes * f_yes) / (pi_yes * f_yes + pi_no * f_no)
print(f"probability of issuing dividend given X=4: {p_yes:.4f}")
print(f"value in percentage: {p_yes*100:.2f}%")
```
