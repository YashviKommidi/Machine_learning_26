The following questions are relative to k-fold cross-validation.

  a) Explain how k-fold cross-validation is implemented.

> **_Answer:_**  K fold cross validation is generally implementing by dividing the whole dataset of n observations into k equal sized groups called folds, so thats why its called k fold cross validation. For example if the k = 6, then we will split the whole dataset into 6 roughly equal parts. Then we repeat the whole process k times. that is in each iteration, one fold is taken as validation set and the remaining k-1 folds are combined and used as training set. The model is fitted on the training set and then tested on the validation set which is set aside at the start, then we will record the error for that fold. So we will repeat this whole process for every fold available, as we have known folds, we will repeat this process k times. Each time we will use different fold as the validation set, so that every observation gets a chance to be in the validation set exactly once.

> $$CV_{(k)} = \frac{1}{k} \sum_{i=1}^{k} MSE_i$$

> At last we will take all the errors that is k recorded errors and they are averaged together to produce the overall cross validation error estimate. So the final estimate gives us the good approximation of the test error of the model.  Usually we generally take the k values as 5 or 10, but we can always do the trial and error for selecting k value, but selecting 5 or 10, gives us a good balance between computational cost and accuracy of the error estimate.

  b) How would you choose the analyzed k? Does that matter? (please cite your sources)

> **_Answer:_**  the choice of k value does matter. I feel there is no single correct value of k that would work for every type of data and the models, but there are few guidelines to help to make a better decision in selecting the value. Generally the most commonly recommended values for K is either 5 or 10.  By selecting these values we can say that these values provide a good balance between the computational cost and the accuracy of the error estimate. 



> we can say that the K value directly affects the bias variance tradeoff of error estimate. Which means that a higher k value means that each model is trained on a larger training set and tested on a smaller test fold. In this scenario, this should give us a lower prediction error as the models see more of the available data. But in other scenario , a very large k value such as LOOCV scenario means training many more models which can be computationally expensive. With using the LOOCV the estimates from each fold are highly correlated and hence their average can have high variance. 



> Where as in another scenario of small K value, a lower k means that the model is trained on a smaller training set and tested on a larger test fold. In this case we can expect a higher predcition error on average which leads to higher bias in error estimate.



>Usually k is taken 5 or 10 because these provide a good balance between computational efficiency and accuracy of the error estimate. The source which i referred to also suggests that averaging the result over multiple values of K can make results more robust to this choice, as single value of K can be heavily affected by random fold splits.



>source is Olsen, L.R. (2025). "Multiple-k: Picking the number of folds for cross-validation." 
 https://cran.r-project.org/web/packages/cvms/vignettes/picking_the_number_of_folds_for_cross-validation.html

  c) What are the advantages and disadvantages of k-fold cross-validation relative to *LOOCV*?

> **_Answer:_**  when comparing the K fold cross validation to LOOCV we can point out advantages and disadvantages such as 

> -> k fold cross validation is much less computationally expensive than the LOOCV, where as LOOCV we need to fit the model n times whereas in Kfold cross validation we only fit the model k times. So when n value is large, LOOOCV can be very slow and expensive. The estimates from each fold with LOOCV are highly correlated since each training set is almost identical, differing by only one observation. Which means that their average can have high variance, K fold CV produces less correlated estimates and therefore a lower variance error estimate.



> -> coming to the disadvantages, we can list few such as since k fold cross validation uses fewer observations for training compared to LOOCV, it tends to have higher bias in the error estimate where as LOOCV trains on n-1 observations each time which is almost the entire dataset giving nearly unbiased estimate. In the scenario of small datasets, LOOCV is easy as n value is low, as its uses almost all the data for training in each iteration, where as k fold cross validation may not have enough observations in each fold to train the model effectively.

  d) What are the advantages and disadvantages of k-fold cross-validation relative to *the validation set* approach?

> **_Answer:_** when we are comparing the k fold cross validation with the validation set we can point out few advantages and disadvantages such as the validation set approach has major drawback in that the test error estimate can be highly variable which means that the results can change a lot depending on which observations that are randomly chosen for training and which are chosen for validation. In the case of K fold cross validation, this is fixed by the making sure every observation gets a turn in the validation set which gives us a much more stable and reliable estimate of the test error. 

> in the validation set approach, only a portion of the data is used for training, the model is fitted on fewer observations. This means that the validation error tends to overestimate the true test error. K fold cross validation uses more data for training  that is in each iteration, which gives more accurate estimate of the test error. 

> on the other hand when it comes to disadvantages, K fold cross validation is more computationally expensive than the validation set approach since it requires fitting the model K  times instead of just once. we can generally say that in the case of speed and simplicity, the validation set approach may be preferred since it is faster and easier to implement.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.preprocessing import PolynomialFeatures
import statsmodels.api as sm

# Simulate some data
np.random.seed(0)
n = 100
order = 2
x = np.random.uniform(-4, 4, n)
y = np.random.normal(loc=x**order, scale=0.25, size=n)
data = pd.DataFrame({'x': x, 'y': y})

# Number of folds
n_folds = 20

# Polynomial degrees
poly_degrees = range(1, 11)

# KFold cross-validation
kf = KFold(n_splits=n_folds, shuffle=True, random_state=1)
mse_estimates = []

# Loop over polynomial degrees
for q in poly_degrees:
    y_hat = np.zeros(n)
    
    for train_idx, test_idx in kf.split(data):
        # Fit on K-1 folds
        poly = PolynomialFeatures(degree=q)
        X_train = poly.fit_transform(data.iloc[train_idx][['x']])
        X_test = poly.transform(data.iloc[test_idx][['x']])
        
        ols = sm.OLS(data.iloc[train_idx]['y'], X_train).fit()
        
        # Predict on the Kth fold
        y_hat[test_idx] = ols.predict(X_test)
    
    # Get the MSE estimate
    mse = np.mean((data['y'] - y_hat) ** 2)
    mse_estimates.append(mse)
```

> **_Answer:_**
from the code we can see that it is implementing the K fold cross validation to compare the polynomial regression models of different degrees and out of that find the best one with the help of mean squared error (MSE) which is the last step.



> firstly from the code we can see that it is generating a random dataset of n =100 observations and then x values are randomly drawn from a uniform distribution between the -4 and 4 then the y values are generated base on a quadratic relationship with some random noise added. This generally means that the true underlying pattern in the data is a degree 2 polynomial. 

> Next we can see that a value of 20 has been assigned as k value, which means it is a 20 fold cross validation using the KFold function, and also it defines a range of polynomial degrees from 1 to 10 that it wants to test.  Next we can see that for each polynomial degree from 1 to 10, the code runs the full K fold cross validation and in each fold, it fits a polynomial regression model of degree q on the K-1 training folds and then predicts on the held out validation fold. 



> As it is for 20 folds, this process will be repeated for 20 folds so that every observation gets a chance to be in the validation set exactly once.

> Atlast, after all the folds are completed for a polynomial degree, the code then calculates the MSE values by comparing the predicted values to the actual y values. This MSE values is stored in the mse_estimates list.



> we can say that the whole code is to use K fold cross validation to estimate the test error for polynomial regression models of degrees 1 to 10, so that we can identify which polynomial degree fits the data best. 

b) To the best of your knowledge, improve the structure, content, clarity, and reproducibility of the code presented before in part a) of this question (e.g. would you run a single or multiple replicates?). Fix any mistakes (if you find any). Finally, generate at least two plots summarizing your findings regarding the best-fitting polynomial order on the simulated dataset (also from part a): (1) MSE vs polynomial order, and (2) x vs y, along with a plot of the selected model.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
from sklearn.preprocessing import PolynomialFeatures
import statsmodels.api as sm

# Simulate some data
np.random.seed(0)
n = 100
order = 2
x = np.random.uniform(-4, 4, n)
y = np.random.normal(loc=x**order, scale=0.25, size=n)
data = pd.DataFrame({'x': x, 'y': y})

# Number of folds and also multiple replicates
n_folds = 10 #changed the no of folds to 10 to get better bias variance tradeoff.
n_replicates = 10  # running multiple replicates to get more stable MSE estimates

# Polynomial degrees
poly_degrees = range(1, 11)

#storing MSE for each replicate and degree
all_mse = np.zeros((n_replicates, len(poly_degrees)))



#looping over replicates
for rep in range(n_replicates):
    #Kfold cross-validation with different random state per replicate
    kf = KFold(n_splits=n_folds, shuffle=True, random_state=rep)

    # Looping over polynomial degrees
    for i, q in enumerate(poly_degrees):
        y_hat = np.zeros(n)
        for train_idx, test_idx in kf.split(data):
            #fitting on K-1 folds
            poly = PolynomialFeatures(degree=q)
            X_train = poly.fit_transform(data.iloc[train_idx][['x']])
            X_test = poly.transform(data.iloc[test_idx][['x']])
            ols = sm.OLS(data.iloc[train_idx]['y'], X_train).fit()
            #predicting on the Kth fold
            y_hat[test_idx] = ols.predict(X_test)
        # getting the MSE estimate
        all_mse[rep, i] = np.mean((data['y'] - y_hat) ** 2)
        
        
#finding the average MSE across all the replicates
mse_estimates = all_mse.mean(axis=0)
best_degree = list(poly_degrees)[np.argmin(mse_estimates)]
print(f"best polynomial degree is : {best_degree}")


#creating a plot between MSE vs polynomial degree
plt.figure(figsize=(8, 5))
plt.plot(list(poly_degrees), mse_estimates, marker='o', color='steelblue', linewidth=2)
plt.axvline(x=best_degree, color='red', linestyle='--', label=f'best degree = {best_degree}')
plt.xlabel('polynomial degree')
plt.ylabel('cross validation MSE')
plt.title('MSE vs polynomial degree')
plt.xticks(list(poly_degrees))
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
plt.close()

# creating a plot between x vs y with best fitting model
poly_best = PolynomialFeatures(degree=best_degree)
X_full = poly_best.fit_transform(data[['x']])
ols_best = sm.OLS(data['y'], X_full).fit()
x_plot = np.linspace(-4, 4, 300)
y_plot = ols_best.predict(poly_best.transform(pd.DataFrame(x_plot, columns=['x'])))

plt.figure(figsize=(8, 5))
plt.scatter(data['x'], data['y'], alpha=0.5, color='steelblue', s=30, label='data')
plt.plot(x_plot, y_plot, color='red', linewidth=2, label=f'best model (degree={best_degree})')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'x vs y with best fitting polynomial (degree={best_degree})')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
plt.close()
```

firstly i have changed the value of k to get better bias variance tradeoff. Then I have added 10 replicates, each with a different random seed and averaged the MSE across all replicates to get a more stable and reliable estimate of the test error as the original code only has a single replicate of K fold CV which generally means the results could be heavily affected by a single random fold split. 

and I also noticed than k fold was defined outside the degree loop with a fixed random state, which means that every degree used the exact same fold splits. So i used that inside the replicate loop so each replicate uses a different random seed. 



At last I have added 2 visualizations that is a plot between x vs y with best fitting model and a plot between MSE vs polynomial degree to identify the best degree and to visually confirm the result.



#### Question 4

Bootstrap the following dataset (`n = 1000`) to obtain median and 95% CI for parameter estimates (slope and intercept) summarizing the relationship between `x` and `y_measured.` What happens with median parameter estimates when you examine `y` instead? Provide some visualizations to explain your results.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


np.random.seed(42)
nobs = 1000
x = np.random.normal(size=nobs)
y = 5 - 2.25 * x
y_measured = y + np.random.normal(size=nobs)


#writing the bootstrap function
def bootstrap_regression(x, y, n_bootstrap=1000):
    slopes, intercepts = [], []
    for _ in range(n_bootstrap):
        idx = np.random.choice(len(x), size=len(x), replace=True)
        model = LinearRegression().fit(x[idx].reshape(-1, 1), y[idx])
        slopes.append(model.coef_[0])
        intercepts.append(model.intercept_)
    return np.array(slopes), np.array(intercepts)

#running the bootstrap function for both the y_measured and y
np.random.seed(42)
slopes_m, intercepts_m = bootstrap_regression(x, y_measured)
slopes_y, intercepts_y = bootstrap_regression(x, y)

# printing the results
for name, slopes, intercepts in [("y_measured", slopes_m, intercepts_m), ("y (no noise)", slopes_y, intercepts_y)]:
    print(f"\n{name}")
    print(f"slope: median = {np.median(slopes):.4f}, 95% CI = [{np.percentile(slopes, 2.5):.4f}, {np.percentile(slopes, 97.5):.4f}]")
    print(f"intercept: median = {np.median(intercepts):.4f}, 95% CI = [{np.percentile(intercepts, 2.5):.4f}, {np.percentile(intercepts, 97.5):.4f}]")

# creating first plot on bootstrap distributions with slope and intercept for y_measured
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# code for slope plot
axes[0].hist(slopes_m, bins=40, color='steelblue', alpha=0.7)
axes[0].axvline(np.median(slopes_m), color='red', linestyle='--', label=f'median = {np.median(slopes_m):.4f}')
axes[0].axvline(np.percentile(slopes_m, 2.5), color='green', linestyle=':', label='95% CI')
axes[0].axvline(np.percentile(slopes_m, 97.5), color='green', linestyle=':')
axes[0].axvline(-2.25, color='orange', linestyle='-', label='true = -2.25')
axes[0].set_title('bootstrap slope (y_measured)')
axes[0].set_xlabel('slope')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# code for intercept plot
axes[1].hist(intercepts_m, bins=40, color='steelblue', alpha=0.7)
axes[1].axvline(np.median(intercepts_m), color='red', linestyle='--', label=f'median = {np.median(intercepts_m):.4f}')
axes[1].axvline(np.percentile(intercepts_m, 2.5), color='green', linestyle=':', label='95% CI')
axes[1].axvline(np.percentile(intercepts_m, 97.5), color='green', linestyle=':')
axes[1].axvline(5, color='orange', linestyle='-', label='true = 5')
axes[1].set_title('bootstrap intercept (y_measured)')
axes[1].set_xlabel('intercept')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# creating second plot which is scatter plots for x vs y_measured and x vs y
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
x_plot = np.linspace(x.min(), x.max(), 300)

# code for plot x vs y_measured
model_m = LinearRegression().fit(x.reshape(-1, 1), y_measured)
axes[0].scatter(x, y_measured, alpha=0.3, color='steelblue', s=10, label='data')
axes[0].plot(x_plot, model_m.predict(x_plot.reshape(-1, 1)), color='red', linewidth=2, label=f'slope={model_m.coef_[0]:.2f}, intercept={model_m.intercept_:.2f}')
axes[0].set_title('x vs y_measured')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

#code for plot x vs y (no noise)
model_y = LinearRegression().fit(x.reshape(-1, 1), y)
axes[1].scatter(x, y, alpha=0.3, color='salmon', s=10, label='data')
axes[1].plot(x_plot, model_y.predict(x_plot.reshape(-1, 1)), color='red', linewidth=2, label=f'slope={model_y.coef_[0]:.2f}, intercept={model_y.intercept_:.2f}')
axes[1].set_title('x vs y (no noise)')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

```

We will predict the number of applications received using the other variables in the `College` data set. Please load the relevant dataset first.

  a) Split the data set into a training set and a test set (70%, 30%).

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

#loading the College.csv dataset
college_data = pd.read_csv('College.csv', index_col=0)

# converting Private column to 0 and 1
college_data['Private'] = (college_data['Private'] == 'Yes').astype(int)

# setting up the X and y
X = college_data.drop(columns='Apps')
y = college_data['Apps']

#splitting the data into training and test set that is 70/30.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(f"training size: {X_train.shape[0]}, test size: {X_test.shape[0]}")
```

  b) Fit a linear model using least squares on the training set, and report the test error obtained.

```python
#importing libraries
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# fitting the linear regression model on training set
ols = LinearRegression()
ols.fit(X_train, y_train)
ols_preds = ols.predict(X_test)
ols_mse = mean_squared_error(y_test, ols_preds)
print(f"linear regression test MSE: {ols_mse:.2f}")
```

by using the above code we obtained the test MSE value on the held out test set is 1931803.19

  c) Fit a lasso model on the training set, with $\lambda$ chosen by cross-validation. Report the test error obtained, along with the number of non-zero coefficient estimates.

```python
#importing library
from sklearn.linear_model import LassoCV

#scaling the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# fitting the lasso with lambda chosen by cross validation
lasso = LassoCV(cv=10, random_state=42, max_iter=10000)
lasso.fit(X_train_scaled, y_train)
lasso_preds = lasso.predict(X_test_scaled)
lasso_mse = mean_squared_error(y_test, lasso_preds)
non_zero = np.sum(lasso.coef_ != 0)
print(f"lasso test MSE: {lasso_mse:.2f}")
print(f"best lambda value : {lasso.alpha_:.4f}")
print(f"number of non-zero coefficients : {non_zero}")
```

here we have fit a lasso model on the training set with λ choosen by 10 fold cross validation. The best value for the λ was 3.7842 and the test MSE value obtained was 1927762.16 and the number of non zero coefficients is 17 which generally means that the Lasso kept all the predictors in the model. compared to the linear regression test MSE, lasso performed slightly better.

  d) Fit a ridge regression model on the training set, with $\lambda$ chosen by cross-validation. Report the test error obtained.

```python
#importing required library
from sklearn.linear_model import RidgeCV

# fitting the ridge regression with the lambda chosen by cross validation
ridge = RidgeCV(alphas=np.logspace(-3, 6, 100), cv=10)
ridge.fit(X_train_scaled, y_train)
ridge_preds = ridge.predict(X_test_scaled)
ridge_mse = mean_squared_error(y_test, ridge_preds)
print(f"ridge regression test MSE value : {ridge_mse:.2f}")
print(f"best lambda value: {ridge.alpha_:.4f}")
```

Here  with this code I have fit a ridge regression model on the training set .The best λ value was 2.3101 and on the other hand the test MSE I got is 1863395.23. when we compare this to both linear regression  and Lasso, we can see that ridge regression performs the best among the three models.

#### Question 6

What is the best polynomial order (1–20) modeling the structure of the following dataset (`x` vs `y`)? Your answer should include a numerical approach (e.g. MSE estimates) along with a convincing set of visuals supporting your answer.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(4)
n = 1000
x = np.random.uniform(-4, 4, n)
mu = np.where(x > 0, np.cos(4 * x), 1 - np.sin(x))
y = mu + np.random.normal(size=n)
data = pd.DataFrame({'x': x, 'y': y})

plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='black', edgecolor='steelblue', s=50, label='Data', alpha=0.7)
x_ordered = np.sort(x)
mu_ordered = mu[np.argsort(x)]

plt.plot(x_ordered, mu_ordered, color='red', linewidth=3, label="E[Y|X=x], truth")

plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc='lower left')
plt.show()
```

continung the code with the help of previous code

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error

# using the 10-fold Cross validation for degrees 1 to 20
X = x.reshape(-1, 1)
kf = KFold(n_splits=10, shuffle=True, random_state=42)
degrees = list(range(1, 21))

# creating lists to store average MSE across folds for each degree
cv_mse = []
train_mse = []

for deg in degrees:
    fold_val_errors = []
    fold_train_errors = []

    for train_idx, val_idx in kf.split(X):
        #splitting the data into training and validation sets 
        X_train, X_val = X[train_idx], X[val_idx]
        y_train, y_val = y[train_idx], y[val_idx]
        
        # building the  pipeline 
        model = Pipeline([('poly', PolynomialFeatures(degree=deg)), ('lr', LinearRegression())])
        model.fit(X_train, y_train)

        fold_val_errors.append(mean_squared_error(y_val, model.predict(X_val)))
        fold_train_errors.append(mean_squared_error(y_train, model.predict(X_train)))

    cv_mse.append(np.mean(fold_val_errors))
    train_mse.append(np.mean(fold_train_errors))


cv_mse = np.array(cv_mse)
train_mse = np.array(train_mse)

#selecting the best degree
best_deg = degrees[np.argmin(cv_mse)]
print(f"best degree: {best_deg}, cross validation MSE: {cv_mse[best_deg-1]:.4f}")

# plotting the first plot between train MSE vs cross validation MSE
plt.figure(figsize=(8, 5))
plt.plot(degrees, train_mse, 'o--', label='train MSE')
plt.plot(degrees, cv_mse, 's-', label='cross validation MSE')
plt.axvline(best_deg, color='red', linestyle='--', label=f'best degree = {best_deg}')
plt.xlabel('polynomial degree')
plt.ylabel('MSE')
plt.title('train vs cross validation MSE by polynomial degree')
plt.xticks(degrees)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

#plotting the second plot between fitted curve for best degree vs true function
x_plot = np.linspace(-4, 4, 600).reshape(-1, 1)
best_model = Pipeline([('poly', PolynomialFeatures(degree=best_deg)), ('lr', LinearRegression())])
best_model.fit(X, y)

plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='black', edgecolor='steelblue', s=50, alpha=0.4, label='data')
plt.plot(x_ordered, mu_ordered, color='red', lw=3, label='True E[Y|X]')
plt.plot(x_plot, best_model.predict(x_plot), color='green', lw=2, label=f'Degree {best_deg} fit')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'best polynomial fit (degree {best_deg}) ')
plt.legend()
plt.show()
```

here in this code, to identify the best polynomial order, i have evaluated the models from degree 1 to 20 using the 10 fold cross validation. So in each iteration, the data was split into 10 folds where 9 were used for the training and 1 for the validation and the MSE was averaged across all the folds. So this gives us a more reliable estimate of test error compared to a single train and validation split. 



If we look at the cross validation MSE plot, we can see that the error steadily decreases from the degree 1 where it was 1.534 and it went down to degree 17 that is MSE value is 0.981, after this point it began to rise slightly for the degrees 18-20. We can say that this is a sign of overfitting that is the model starts fitting the noise rather than the underlying pattern. The other one , that is the training MSE graph keeps decreasing all the way to degree 20 which is usually expected as more complex model always fits the training data better. 



By looking at these results we can say that degree 17 is selected as the best polynomial order as it achieves the lowest cross validation MSE. When it comes to the next plot, it gives us a clarity why 17 is best, that is the degree 17 fitted curve closely follows the true E[Y|X] function without overfitting to the noise in the data.
