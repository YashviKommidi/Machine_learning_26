a) Discuss the implications that uncertainty in $X$ would have on parameter estimates based on linear regression models. Should we care about error in $X$ when fitting OLS regressions? 

> **_Answer:_** Generally in linear regression, we usually think that $X$ is measured correctly and all the error comes from $ε$. But actually what can happen is, X can have errors too when we measure it.So when X has errors in it, this can also affects the coefficient estimates from the OLS regression. 

> Here the estimates become biased, meaning that they are not close to the true values. This problem is generally called as regression dilution. What happens here is that the slope $β$₁ gets pulled towards the zero and it ends up being smaller than the true value. So basically, the model makes the relationship between X and y look weaker than it really is.

>And yes, we should care about the errors in X when we are performing the OLS regression. If we just ignore the uncertainty in X, then there is a chance that we will get wrong results. Our parameter estimates will be biased, the confidence intervals will not be correct, and we will not understand the true relationship properly. And also, we can say that more error in X means more bias in our estimates.

b) Provide a brief discussion on the concept of regression dilution.

> **_Answer:_** Regression dilution is a problem that generally happens when there is a measurement error in the predictor variable $X$. When $X$ is measured with error, the estimated slope $β₁$ gets biased towards zero and becomes smaller than the true value. So basically, this makes the whole relationship between $X$ and $y$ appear to be weaker than it actually is.

>We can say that the actual reason why this happens is because the measurement error in $X$ adds extra noise which makes it harder for the model to correctly identify what the true relationship is. To understand this in a better way lets consider a case, if the true slope should be equal to value 0.8, but X has few measurement errors, then the estimated slope might come out as 0.5 or 0.6. So the model underestimates how strong the relationship really is between $X$ and $y$.

>We can also say that the more measurement error there is in $X$ variable, the worse this dilution effect becomes. If we see in general real world cases, this can be a serious issue in areas like medicine where measurements are often not that perfect. So to fix this problem, we either need to measure the value of $X$ more accurately or use special regression methods which can handle the measurement errors, like errors-in-variables models or total least squares regression.

c) Are there any linear regression models that account for uncertainty in $X$ and $Y$? Please do some independent research and cite your sources.

> **_Answer:_** Yes, there are several linear regression models that can account for uncertainty in both X and Y variables. After going through a paper, I understood about this in a better way.

>Total Variance (TV) method: This method is generally considered as one of the best approaches, because it generally works by minimizing the sum of weighted squared errors in both $X$ and $Y$ variables. This method can work for any response function, whether it is linear or nonlinear, and it requires estimates of the data errors for both the variables. Next  is the Effective variance (EV2) method, where this is a modified version of the effective variance approach where the uncertainties from all variables are projected into a single quantity. So for the linear response functions, it usually gives the results that are identical to the TV method and it is easy to implement in standard programs. One more model is Deming regression which is a method that is commonly used in analytical chemistry and methods comparison studies.  At last, I would also like to mention about the orthogonal regression, where this method is generally used to minimize the sum of the squared perpendicular distances from the points to the fitted line. We can say that basically, these methods are more complex than the ordinary least squares because these adjust both X and Y values to minimize the total error rather than just minimizing the error in Y alone.

>Below is the source I used to refer and learn about these: Tellinghuisen, J. (2020). Least Squares Methods for Treating Problems with Uncertainty in x and y. Analytical Chemistry, 92, 10863-10871. https://dx.doi.org/10.1021/acs.analchem.0c02178

d) Distinguish between least squares and maximum likelihood in the context of linear regression models.

> **_Answer:_** In the context of linear regression models, least squares and maximum likelihood methods are two different approaches that are generally used to estimate the coefficients.

> First, least squares method generally works by minimizing the sum of squared differences between the actual values and also the predicted values. So basically this means that, it tries to find the line that is closest to all the data points by making the squared errors as small as possible. Also, in this method, the model does not make any specific assumptions about the distribution of the errors, it just focuses on minimizing the squared residuals. On the other hand, maximum likelihood usually works in a different way that is, it finds the coefficient values that make the observed data most likely to have occurred. Here, this method assumes that the error term follows a particular way of distribution, which is usually the normal distribution with mean value 0 and some variance. So here, it tries to find the coefficients that maximize the probability of seeing the data we actually have.



> One more point we can talk about is, when we assume the errors are normally distributed, both the methods give us exactly the same coefficient estimates. This is why ordinary least squares method is so commonly used in practice, because it is usually simpler to compute and it gives us the same results as maximum likelihood under the normality assumption.

> However, there are some differences that we need to look at between them. Maximum likelihood also estimates the variance of the error term, while least squares does not always do this. And also, we can say that the maximum likelihood is more flexible because it can be used with different error distributions, while least squares is specifically designed to minimize the squared errors.

This question involves the use of simple linear regression on the `Auto` data set.

a) Use the `statsmodels.api` package to perform a simple linear regression with `mpg` as the response and `horsepower` as the predictor. Print the regression results. 


```python
import pandas as pd
#using the statsmodels.api package
import statsmodels.api as sm
import numpy as np

# Loading the Auto dataset
auto_data = pd.read_csv('Auto.data', sep=r'\s+', na_values='?')

# removing the rows with missing values
auto_clean = auto_data.dropna(subset=['horsepower', 'mpg'])

# selecting the variables for regression
X = auto_clean['horsepower']  # horsepower as predictor
y = auto_clean['mpg']          # mpg as response

# adding the constant that is intercept to the model
X_with_const = sm.add_constant(X)

# fitting the simple linear regression model
model = sm.OLS(y, X_with_const)
results = model.fit()
print(f"RSE: {np.sqrt(results.mse_resid):.4f}")

# printing the regression results we got
print(results.summary())
```

#### Question 8

It is claimed that in the case of simple linear regression of $Y$ onto $X$, the $R^2$ statistic is equal to the square of the correlation between $X$ and $Y$. Generate some synthetic data and show that this is the case.

```python
import numpy as np
import statsmodels.api as sm

# generating the synthetic data
np.random.seed(42)
n = 100
X = np.random.uniform(0, 10, n)
Y = 2 + 3*X + np.random.normal(0, 2, n)

# calculating the R² from regression
X_with_const = sm.add_constant(X)
model = sm.OLS(Y, X_with_const)
results = model.fit()
r_squared = results.rsquared

# calculating the correlation squared
correlation = np.corrcoef(X, Y)[0, 1]
correlation_squared = correlation**2

# printing the values of R² and Correlation²
print(f"R² from regression: {r_squared:.6f}")
print(f"Correlation²: {correlation_squared:.6f}")

#printing the difference
print(f"Difference: {abs(r_squared - correlation_squared):.10f}")

# creating a scatter plot with regression line to visualise the difference and understand it better
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, alpha=0.6, label='data points')
plt.plot(X, results.predict(X_with_const), color='red', linewidth=2, label='regression line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('simple linear regression')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
